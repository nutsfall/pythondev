## 仕様

この課題は「2次元リストを作成して、そのリストから値を取り出して表示する」という理解を確認する内容です。

以下の仕様に沿って、2つのコードセルの追記・修正を行うこと。

---

### 1. 九九のリストを作成する

- 「#@title 九九のリストを作成する」と書かれた行の下にコードを追加する  
- 九九（1〜9）の結果を格納した 2次元リスト を作成する  
- 作成方法は以下のいずれかを使用すること  
  - リスト内包表記  
  - 二重ループ  
- **直書きによる 9×9 のリスト定義は禁止**

---

### 2. 九九の表を表示する

- 「#@title 九九の表を表示する」と書かれた行の下にある 二重の for 文を修正する
- 作成した2次元リストから値を取り出して九九を表示する  
- 出力形式は、指定された実行結果と同じレイアウトにすること  

```python
#@title 九九のリストを作成する
print([[i * j for j in range(1, 10)] for i in range(1, 10)])
```

```python
#@title 九九の表を表示する
cells81 = [[i * j for j in range(1, 10)] for i in range(1, 10)]
for v in range(1, 10):
    for h in range(1, 10):
        print(f"{cells81[v-1][h-1]:3d}", end="")

    print("")
```
