## 仕様

- `range` と繰り返し処理を使って、1から100までの連続する整数値を画面に表示する
- ただし、素数でないものは表示しない

### 素数判定のルール

以下のルールに従って素数かどうかを判定する：

- 1は素数ではない
- 2を除いた2の倍数は素数ではない
- 3を除いた3の倍数は素数ではない
- 5を除いた5の倍数は素数ではない
- 7を除いた7の倍数は素数ではない

### 実装方針

- 上記の判定手順に基づき、エラトステネスのふるいを用いて処理を実装する

```python
for i in range(1,101):
    
    if i == 1:
        continue

    for j in [2,3,5,7]:
        if i != j and i % j == 0:
            break
    else:
        print(i)
```
